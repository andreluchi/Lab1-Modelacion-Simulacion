```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_campo_direcciones(f, xmin, xmax, ymin, ymax, x_puntos, y_puntos, tipo_campo='F', mostrar_lineas_flujo=False):
    """
    Grafica el campo de direcciones para una ecuación diferencial definida por f(x, y).
    Parámetros:
    - f: función que define la ecuación diferencial dy/dx = f(x, y)
    - xmin, xmax, ymin, ymax: límites 
    - x_puntos, y_puntos: número de puntos en cuadricula en los ejes x e y
    - tipo_campo: 'F' para el campo normal, 'N' para el campo normalizado
    - mostrar_lineas_flujo: si se deben agregar líneas de flujo al gráfico
    """
    x = np.linspace(xmin, xmax, x_puntos)
    y = np.linspace(ymin, ymax, y_puntos)
    X, Y = np.meshgrid(x, y)
    U, V = np.ones_like(X), f(X, Y)
    
    if tipo_campo == 'N':
        magnitud = np.sqrt(U**2 + V**2)
        U, V = U / magnitud, V / magnitud

    plt.figure(figsize=(10, 8))
    plt.quiver(X, Y, U, V, color='blue', angles='xy')
    
    if mostrar_lineas_flujo:
        plt.streamplot(X, Y, U, V, color='red')
    
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Campo de Direcciones {"N" if tipo_campo == "N" else ""}')
    plt.grid(True)
    plt.show()


# AQUI SE DEFINE LA ECUACION DIFERENCIAL POR FAVOR CAMBIARLA SEGUN SEA NECESARIO
def ecuacion_diferencial(x, y):
    return y + x

def main():
    # Parámetros de visualización
    xmin, xmax = -10, 10
    ymin, ymax = -10, 10
    x_puntos, y_puntos = 20, 20  # Número de puntos en cuadricula

    # inputs
    tipo_campo = input("Elija el tipo de campo (F o N): ").strip().upper()
    while tipo_campo not in ['F', 'N']:
        print("Entrada inválida. Por favor, ingrese 'F' o 'N'")
        tipo_campo = input("Elija el tipo de campo (F o N): ").strip().upper()

    mostrar_lineas_flujo = input("¿Desea mostrar líneas de flujo? (si/no): ").strip().lower() == 'si'

    # Graficar el campo de direcciones según las elecciones del usuario
    graficar_campo_direcciones(ecuacion_diferencial, xmin, xmax, ymin, ymax, x_puntos, y_puntos, tipo_campo, mostrar_lineas_flujo)


```

Discontinuidad en: 𝑥=0

lo que significa que el teorema de existencia y unicidad no se aplica en 𝑥=0.


En x=0, debido a la discontinuidad, pueden ocurrir tres cosas:

No hay solución: Si la condición inicial implica una evaluación en 𝑥=0 para 𝑦′, la solución no existe.
Múltiples soluciones: Sin una condición inicial específica que evita 𝑥=0, pueden existir múltiples soluciones, dependiendo de cómo se resuelve la singularidad.
Solución única fuera de 𝑥=0: Si la condición inicial está definida para 𝑥≠0, puede haber una solución única en esa región.

```python
import numpy as np
import matplotlib.pyplot as plt


def y1(x):
    return x**2 + 1  # Para y(1) = 2

def y2(x):
    return x**2 - 3  # Para y(1) = -2

def y3(x):
    return x**2  # Para y(1) = 1


x = np.linspace(-10, 10, 400)


plt.figure(figsize=(12, 8))
plt.plot(x, y1(x), label='y = x^2 + 1')
plt.plot(x, y2(x), label='y = x^2 - 3')
plt.plot(x, y3(x), label='y = x^2')


plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Soluciones de la EDO')
plt.legend()
plt.grid(True)
plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.show()
```
